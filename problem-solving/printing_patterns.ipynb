### **Problem_1:**
** a right triangle shaped as half of 5x5 square. **

```python
n = 15
for i in range(0,n):
  m = n - i
  for i in range(0,m):
    print('#', end='')
  print()
```

```python
# Now for reverse traingle modifying the above code
n = 15
for i in range(0,n):
  # m = n - i
  m = i + 1
  for i in range(0,m):
    print('#', end='')
  print()
```
