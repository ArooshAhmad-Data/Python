A tuple uses less memory than a list of the same length, and it allows Python to do some optimizations.

Once a tuple is created, you cannot change, add, or remove elements from it. Its structure is fixed.<br>
          
*   **But what's inside the tuple can bemutable!**

Tuples can contain mutable objects, like lists or dictionaries:

**Tuple unpacking**
Unpacking refers to how we can unpack values from sequences in python

```python
lax_coordinates = (33.9425, -118.408056)
latitude, longitude = lax_coordinates # unpacking
latitude, longitude
```

```python
a, b, *rest = range(5) # * makes the last rest variable have all the values
a,b, rest
```

```python
a, *rest, b = range(5) # last value is stored into 'b' first into 'a' all others in 'rest'
a, rest, b
```

### Nested Unpacking

```python
metro_areas = [
 ('Tokyo', 'JP', 36.933, (35.689722, 139.691667)),
 ('Delhi NCR', 'IN', 21.935, (28.613889, 77.208889)),
 ('Mexico City', 'MX', 20.142, (19.433333, -99.133333)),
 ('New York-Newark', 'US', 20.104, (40.808611, -74.020386)),
 ('São Paulo', 'BR', 19.649, (-23.547778, -46.635833)),
]
```

```python
def main():
 print(f'{"":15} | {"latitude":>9} | {"longitude":>9}')
 for name, _, _, (lat, lon) in metro_areas:
  if lon <= 0:
    print(f'{name:15} | {lat:9.4f} | {lon:9.4f}')
```

```python
main()
```
