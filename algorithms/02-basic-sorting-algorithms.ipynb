## Selection Sort

Selection works on comparison basis

Comparing and finding an element smallest or largest and
* appending it to new array
* popping from the old array

The new array will have data sorted.

It is same like the music example from grokking algorithms, second edition.

```python
def findSmallest(arr):
  """
    return the index of the smallest element in the array.
  """
  smallest = arr[0]
  smallest_index = 0
  for i in range(1, len(arr)-1):
    if arr[i] < smallest:
      smallest = arr[i]
      smallest_index = i # index of smallest element
  return smallest_index

def selectionSort(arr):
  """
    sort the array using selection sort.
  """
  new_arr = []
  copied_arr = list(arr)
  for i in range(len(arr)):
    smallest = findSmallest(copied_arr) # get index of the smallest element
    new_arr.append(copied_arr.pop(smallest)) # pop that smallest from the copied array and append to new array

  return new_arr
```

```python
data = [5, 3, 6, 2, 10]
print(selectionSort(data))
```
